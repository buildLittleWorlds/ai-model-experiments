## Part 3: Bridge to Level 3 (1:10-1:25)

### You Have Been Doing Research All Semester

Think about what you did:
- You **formed hypotheses** ("I think the Twitter model will handle slang better")
- You **designed experiments** (paired-sentence bias tests, domain shift comparisons)
- You **collected evidence** (model outputs, confidence scores, comparison tables)
- You **drew conclusions** ("the model's training data explains its blind spots")
- You **documented results** (your project README, your observation tables)

That is the scientific method. You have been doing research.

### What Level 3 Looks Like

In Level 2, the question was: **"Try this model and see what happens."**

In Level 3, the question will be: **"What do YOU want to investigate?"**

Instead of following guided experiments, you will:
- **Choose a research question** about AI
- **Design your own experiments** to answer it
- **Collect and analyze data** systematically
- **Write up your findings** like a real researcher
- **Present your research** to the group

### Example Research Questions

These are the kinds of questions Level 3 students might investigate:

- "Do sentiment models treat different languages or dialects fairly?"
- "How much does training data size affect model accuracy?"
- "Can zero-shot classification replace fine-tuned models for specific tasks?"
- "How do different prompting strategies affect text generation quality?"
- "What happens when you chain 5+ models together -- does error compound predictably?"
